```python
from IPython.display import HTML
from data import generator, documents, distributions
```

```python
testset = generator.generate_with_langchain_docs(documents, 20, distributions, with_debugging_logs=True)
```

```python
test_df = testset.to_pandas()
```

```python
test_df
```

```python
test_questions = test_df["question"].values.tolist()
test_groundtruths = test_df["ground_truth"].values.tolist()
```

```python
#from  src.agents.graph import graph
from src.vectorstore.get import retriever
from src.agents.rag_agent import RagAgent
from langchain_openai import ChatOpenAI

answers = []
contexts = []
llm = ChatOpenAI(model="gpt-4o", temperature=0)
agent = RagAgent(llm, retriever)
chain = agent.get_chain()
for question in test_questions:
  response = chain.invoke({"question" : question })
  answers.append(response["response"].content)
  contexts.append([context.page_content for context in response["context"]])
```

```python
from ragas import evaluate
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    answer_correctness,
    context_recall,
    context_precision,
)

metrics = [
    faithfulness,
    answer_relevancy,
    context_recall,
    context_precision,
    answer_correctness,
]
```

```python
from datasets import Dataset

response_dataset = Dataset.from_dict({
    "question" : test_questions,
    "answer" : answers,
    "contexts" : contexts,
    "ground_truth" : test_groundtruths
})
```

```python
response_dataset.to_pandas()
```

```python
results = evaluate(response_dataset, metrics)
```
